```python
#Calendar File Cleaning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_calendar =pd.read_csv(r"C:\Users\mehul\OneDrive\Desktop\Data Analyst\AirBNB\Australia\calendar.csv")
df_calendar.head()
```

```python
df_calendar.drop(columns=['adjusted_price'], inplace=True)
```

```python
df_calendar.columns
```

```python
#Convert Date Column
df_calendar['date'] = pd.to_datetime(df_calendar['date'], dayfirst=True)
df_calendar['date'].dtype
```

```python
#Clean Price Column
df_calendar['price'] = pd.to_numeric(
    df_calendar['price'].astype(str)
    .str.replace('$','')
    .str.replace(',',''),
    errors='coerce'
)
df_calendar['price'].describe()
```

```python
#Identify and Remove Outliers 
df_calendar[df_calendar['price'] > 2000].head()
df_calendar = df_calendar[df_calendar['price'] < 2000]
df_calendar['price'].describe()
```

```python
df_calendar['available'].unique()
```

```python
df_calendar['available'] = df_calendar['available'].map({'t': True, 'f': False})
```

```python
df_calendar.isnull().sum()
```

```python
#Calculate Listing Metrics
calendar_summary = df_calendar.groupby('listing_id').agg({
    'price':'mean',
    'available':'sum',
    'date':'count'
}).reset_index()
```

```python
#Rename Column
calendar_summary.columns = [
    'listing_id',
    'avg_price',
    'available_days',
    'total_days'
]
```

```python
#Calculate Availability Rate
calendar_summary['availability_rate'] = (
    calendar_summary['available_days'] /
    calendar_summary['total_days']
)
```

```python
#Calculate Booking Rate
calendar_summary['booking_rate'] = 1 - calendar_summary['availability_rate']
```
